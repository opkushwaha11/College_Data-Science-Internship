```python
sample_data = [23,27,25,24,22,26]
population_mean = 25
t_stat,p_value = stats.ttest_1samp(sample_data, population_mean)
print("T-statistic:",t_stat)
print("P-value:",p_value)
```

```python
if p_value < 0.05:
   print("The sample")
else:
    print("The sample")
     
```

```python
x = np.array([10, 23, 38, 46, 57])
y = np.array([2, 4, 6, 8, 10])


correlation_coef = np.corrcoef(x, y)[0,1]


heatmap_data = np.array([[1.0, correlation_coef], [correlation_coef, 1.0]])

ax = sns.heatmap(heatmap_data, annot=True, cmap="coolwarm", fmt=".2f",
                 xticklabels=["x", "y"], yticklabels=["x", "y"])

plt.title("Correlation Heatmap")
plt.show()
```

```python
slope,intercept,r_value,p_value,std_err=stats.linregress(x,y)
predicted_y=slope*x+intercept
plt.scatter(x,y,label='Data')
plt.plot(x,predicted_y,color='red',label='regression Line')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.show
```
