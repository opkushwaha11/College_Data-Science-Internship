```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# creating a sample dataset
np.random.seed(0)
n_samples = 100
n_features = 4
data = np.random.randn(n_samples , n_features)
```

```python
print(data)
```

```python

pca = PCA(n_components = 1) # all data will be in 1-D 
transformed_data = pca.fit_transform(data)

```

```python

inverse_transformed = pca.inverse_transform(transformed_data)
```

```python
#plot the original data and the PCA-reconstructed data
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.scatter(data[:,0], data[:,1], alpha=0.7)
plt.title('PCA-Reconstructed data')


#inverse transformed Data
plt.subplot(1,2,2)
plt.scatter(inverse_transformed[:,0], inverse_transformed[:,1])
plt.tight_layout()
plt.show()
```

# Kmean


```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# number of clusters
k = 5
kmeans = KMeans(n_clusters=k)

#fit the model to the data
kmeans.fit(data)
```

```python
# get the cluster centers and labels
centers = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
# plot the data points and cluster centres
plt.scatter(data[:,0] ,data[:,1], c=labels ,cmap='Accent')
plt.scatter(centers[:,0], centers[:,1],c='blue',marker='P', s=100, label='Cluster Centers')
plt.title('k-means Clustering')
plt.legend()
plt.show()
```
